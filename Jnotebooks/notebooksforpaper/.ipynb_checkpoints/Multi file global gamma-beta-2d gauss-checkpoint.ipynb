# Background Estimation

Background estimation works by plotting log of photons by the delay between them, assuming a poisson distribution of photon arrivals and fitting a line. The plot will contain single molecule bursts however, so a threshold (in microseconds) has to be defined where the fit begins. There are better ways of finding the threshold (TBD) but this will work for now

```python
threshold = 1500
d1.calc_bg(bg.exp_fit, time_s=300, tail_min_us=(threshold))
dplot(d1, hist_bg, show_fit=True)
d2.calc_bg(bg.exp_fit, time_s=300, tail_min_us=(threshold))
dplot(d2, hist_bg, show_fit=True)
d3.calc_bg(bg.exp_fit, time_s=300, tail_min_us=(threshold))
dplot(d3, hist_bg, show_fit=True)
d4.calc_bg(bg.exp_fit, time_s=300, tail_min_us=(threshold))
dplot(d4, hist_bg, show_fit=True)
d5.calc_bg(bg.exp_fit, time_s=300, tail_min_us=(threshold))
dplot(d4, hist_bg, show_fit=True)
d6.calc_bg(bg.exp_fit, time_s=300, tail_min_us=(threshold))
dplot(d4, hist_bg, show_fit=True)
```

S = 1/(1+bg+(1-g)*b*E)

To do this we will need to apply the correction factors found so far, and to we'll also use a dual channel burst search instead of an all photon burst search.

```python
def ESgauss(dataset):
    dataset.E_fitter.fit_histogram(model=mfit.factory_gaussian(), verbose=False, pdf=False)
    params = dataset.E_fitter.params
    Efit = params.to_dict()
    E = Efit['center'][0]
    dataset.S_fitter.fit_histogram(model=mfit.factory_gaussian(), verbose=False, pdf=False)
    params = dataset.S_fitter.params
    Sfit = params.to_dict()
    S = Sfit['center'][0]
    return(E, S)
x = []
y = []
for d in [d1, d2, d3, d4, d5, d6]:
    E, S = ESgauss(d)
    x.append(E)
    y.append(S)    
```

```python
#E = np.concatenate((d1.E[0], d2.E[0], d3.E[0], d4.E[0], d5.E[0], d6.E[0]), axis=0)
#y = np.concatenate((d1.S[0], d2.S[0], d3.S[0], d4.S[0], d5.S[0], d6.S[0]), axis=0)
E=x
model = lmfit.models.ExpressionModel('1 / ( 1 + b * g + ( 1 - g ) * b * E )',
                             independent_vars=["E"], nan_policy = "omit")
params = model.make_params(b=0.5, g=1.0)
fit = model.fit(y, params, E=E)
fit.plot_fit(0)
print(fit.fit_report())
```

```python
reciproS = []
for S in y:
    reciproS.append(1/S)
reciproS
model = lmfit.models.ExpressionModel('E * m + c',
                             independent_vars=["E"], nan_policy = "omit")
params = model.make_params(m=0.5, c=1.0)
fit = model.fit(reciproS, params, E=E)
fit.plot_fit(0)
print(fit.fit_report())
pars = fit.params
c = pars["c"].value
m = pars["m"].value
```

```python
beta = m + c - 1
gamma = (c - 1)/(beta)
print("beta = ", beta)
print("gamma = ", gamma)
```
